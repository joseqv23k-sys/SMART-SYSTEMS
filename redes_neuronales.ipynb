**¿Qué es?**

Las redes neuronales convolucionales (CNN) son un tipo de red neuronal especializada en trabajar con datos con estructura de rejilla, como imágenes. Están formadas por capas que aprenden automáticamente patrones (bordes, texturas, formas) sin que uno tenga que programarlos manualmente.


**¿Como funciona?**

Las redes neuronales convolucionales (CNN) son un tipo especial de red diseñada para trabajar con imágenes, voz o audio. Se componen de tres tipos principales de capas:

**Capa convolucional:** aplica filtros (kernels) que recorren la imagen para detectar características (bordes, texturas, colores). Cada filtro produce un mapa de características. Se usan parámetros como número de filtros, stride (desplazamiento) y padding (relleno) para controlar el tamaño de la salida. Después de cada convolución suele aplicarse la función ReLU para introducir no linealidad. Varias capas convolucionales seguidas forman una jerarquía: las primeras detectan rasgos simples y las últimas combinan estos rasgos en formas u objetos más complejos.

**Capa de agrupación (pooling):** reduce el tamaño de los mapas de características para simplificar el modelo, disminuir parámetros y evitar sobreajuste. No tiene pesos; solo aplica una función de agregación sobre regiones de la imagen. Las más comunes son:

**Agrupación máxima (max pooling):** toma el valor máximo de cada región.

**Agrupación media:** toma el promedio de cada región.

**Capa totalmente conectada:** aplanan (flatten) los mapas de características y los conectan con una red densa. Esta capa combina toda la información extraída para clasificar la entrada. Normalmente usa softmax en la salida para devolver una probabilidad entre 0 y 1 para cada clase (en MNIST hay 10 clases: dígitos del 0 al 9).

En conjunto, una CNN transforma la imagen en valores numéricos, extrae y combina características relevantes y finalmente decide a qué clase pertenece con alta precisión.

**¿Para que sirve?**

Las CNN se usan para reconocer y clasificar patrones en datos con estructura espacial, sobre todo imágenes. Algunas aplicaciones típicas son:

* Reconocimiento de dígitos, rostros y objetos en fotografías.

* Detección de señales de tráfico y peatones en coches autónomos.

* Diagnóstico médico mediante análisis de imágenes (rayos X, resonancias).

* Análisis de vídeo y seguridad (detectar personas o movimientos).

* Reconocimiento de voz y audio cuando se convierten en espectrogramas.

* Procesamiento de texto cuando se representa en forma de secuencia/imagen.

En general, sirven para que una máquina aprenda automáticamente las características importantes de las imágenes y tome decisiones (clasificación, detección, segmentación, etc.) sin que un humano tenga que programar esos patrones.

**Ventajas y Desvantajas**

* Ventajas de las CNN

**Aprenden características automáticamente:** no hay que diseñar descriptores a mano; aprenden bordes, texturas y patrones directamente de los datos.

**Alta precisión en visión por computador:** son el estándar actual en reconocimiento de imágenes, detección de objetos y clasificación.

**Compartición de pesos:** el mismo filtro se aplica en toda la imagen, reduciendo parámetros y carga computacional frente a redes densas.

**Invarianza a traslación y escala:** reconocen patrones aunque estén desplazados, rotados o escalados.

**Jerarquía de características:** las primeras capas detectan detalles simples, las últimas capturan patrones complejos (rostros, letras, objetos completos).

**Versatilidad:** se aplican no solo a imágenes, también a audio, texto, vídeo y señales biomédicas.

* Desventajas de las CNN

**Alto costo computacional:** requieren GPUs potentes y tiempo para entrenarse en datasets grandes.

**Necesitan muchos datos:** para evitar sobreajuste suelen requerir miles o millones de ejemplos.

**Difíciles de interpretar (“caja negra”):** no siempre está claro qué características usan para decidir.

**Sensibles a ataques adversarios:** pequeñas perturbaciones pueden causar errores en las predicciones.

**Consumen memoria:** los modelos grandes ocupan bastante almacenamiento y RAM en la inferencia.

**Dependencia de la calidad de los datos:** aunque extraen características automáticamente, siguen siendo sensibles a ruido, iluminación y resolución.


```python
# --------------------------
# SECCIÓN 1: Importar librerías y cargar dataset
# --------------------------
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from tensorflow import keras

# Cargar dataset MNIST
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Normalizar imágenes
x_train, x_test = x_train / 255.0, x_test / 255.0

# Ajustar dimensiones para CNN
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

print("Tamaño del dataset:", x_train.shape, y_train.shape)

# Visualizar algunas imágenes
plt.figure(figsize=(5,5))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(x_train[i].reshape(28,28), cmap="gray")
    plt.title(f"Etiqueta: {y_train[i]}")
    plt.axis("off")
plt.show()

# --------------------------
# SECCIÓN 2: Definir y entrenar CNN
# --------------------------
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# Entrenar (solo 2 épocas en vivo para ahorrar tiempo)
history = model.fit(x_train, y_train, epochs=2,
                    validation_data=(x_test, y_test))

# Evaluación
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print("\nPrecisión CNN en test:", test_acc)

# Guardar modelo
model.save("cnn_mnist.keras")

# --------------------------
# SECCIÓN 3: Gráficas de entrenamiento
# --------------------------
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title("Precisión del modelo")
plt.xlabel("Épocas")
plt.ylabel("Precisión")
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title("Pérdida del modelo")
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.legend()
plt.show()

# --------------------------
# SECCIÓN 4: Comparación con Regresión Logística
# --------------------------
# Aplanar imágenes (para modelo clásico)
x_train_flat = x_train.reshape(-1, 28*28)
x_test_flat = x_test.reshape(-1, 28*28)

log_model = LogisticRegression(max_iter=1000)
log_model.fit(x_train_flat, y_train)

y_pred = log_model.predict(x_test_flat)
acc_log = accuracy_score(y_test, y_pred)

print("\nPrecisión Regresión Logística:", acc_log)
print("Precisión CNN:", test_acc)

# --------------------------
```
